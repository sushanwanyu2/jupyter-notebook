```python
def is_leap(y):
    return y%4==0 and y%100!=0 or y%400==0
```

```python
def my_sort(l):
    for i in range(len(l)-1):
        for j in range(len(l)-i-1):
            if l[j]>l[j+1]:
                l[j],l[j+1]=l[j+1],l[j]
    return l
```

```python
my_sort([1,4,5,3,7])
```
